```python
def luas_volume(p,l,t):
    luas = 2 * (p * l + p * t + l * t) #2pl + 2pt + 2lt | 2(pl+pt+lt)
    volume = p * l * t
    hasil = print('Luas Balok = ', luas,'\nVolume Balok = ',volume)

    return hasil
```

```python
luas_volume(2,3,4)
```

```python
def diskon(total):
    if total >= 100000:
        diskon = 0.10
    else:
        diskon = 0.05
    potongan = total * diskon
    total_bayar = total - potongan
    hasil = print("Total pembelian : Rp", total,
                  "\nDiskon ", diskon*100, "% : Rp", potongan,
                  "\nTotal bayar : Rp", total_bayar)

    return hasil
```

```python
diskon(1500000)
```

```python
#4. Program mencari bilangan terbesar dari dua input
def bilangan_terbesar(a,b):
    if a > b:
        return a
    elif b > a:
        return b
    else:
        return "Sama"  # Jika sama besar
```

```python
bilangan_terbesar(3,7)
```

```python
#5. Menampilkan bilangan genap dan ganjil
def ganjil_genap(angka):
    if angka % 2 != 0:
        return 'bilangan ganjil'
    else:
        return 'bilangan genap'
```

```python
ganjil_genap(3)
```

```python
#6. Menentukan kelulusan mahasiswa
def kelulusan(nilai):
    if nilai >= 70:
        return "Lulus"
    else:
        return "Tidak Lulus"
```

```python
#7. Konversi suhu dari Celsius
def konversi_suhu(celcius):
    F = (9/5) * celcius + 32
    R = (4/5) * celcius
    
    hasil = print("Suhu dalam Fahrenheit = ", F,
                 "\nSuhu dalam Reamur     = ", R)

    return hasil
```

```python
konversi_suhu(25)
```

```python
#8. Menentukan tahun kabisat
def kabisat(tahun):
    if (tahun % 400 == 0) or (tahun % 4 == 0 and tahun % 100 != 0):
       return "Tahun Kabisat"
    else:
       return "Bukan tahun kabisat"
```

```python
kabisat(1000)
```

```python
#9. Menentukan jenis segitiga
def jenis_segitiga(a,b,c):
    if a == b == c:
        return "Segitiga Sama Sisi"
    elif a == b or a == c or b == c:
        return "Segitiga Sama Kaki"
    else:
        return "Segitiga Sembarang"
```

```python
jenis_segitiga(4,4,4)
```

```python
#10. Pencampuran Warna
def campur_warna(warna1, warna2):
    if (("Merah" in [warna1, warna2]) and ("Kuning" in [warna1, warna2])):
        return "Warna campuran adalah Orange"
    elif (("Kuning" in [warna1, warna2]) and ("Biru" in [warna1, warna2])):
        return "Warna campuran adalah Hijau"
    elif (("Biru" in [warna1, warna2]) and ("Merah" in [warna1, warna2])):
        return "Warna campuran adalah Ungu"
    else:
        return "Lainnya"
```

```python
campur_warna('Merah', 'Kuning')
```
