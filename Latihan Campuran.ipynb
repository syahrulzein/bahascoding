```python
#Soal 1
def prima_tertentu(start, end):
    for i in range(start, end + 1):
        if i > 1:
            for j in range(2, i):
                if (i % j) == 0: #bukan bilangan prima
                    break
            else:
                hasil = print(i)
    return hasil
```

```python
prima_tertentu(2, 20)
```

```python
#Soal 2
def fibonacci(batas):
    angka_1 = 0
    angka_2 = 1
    hasil = []
    for i in range(batas):
        hasil.append(angka_1)
        lanjut = angka_1 + angka_2
        angka_1 = angka_2
        angka_2 = lanjut
    return hasil
```

```python
fibonacci(10)
```

```python
#Soal 3
def index_ganjil(my_list):
    for i in my_list[1::2]: #[a:b:c] -> a = awal, b -> akhir, c -> step
        hasil = print(i, end = ' ')
    return hasil
```

```python
my_list = [1,2,3,4,5,6,7,8]
index_ganjil(my_list)
```

```python
#Soal 4
def deret_bilangan(angka, terms):
    angka_terakhir = angka
    hasil = 0
    for i in range(0, terms):
        print(angka, end='+')
        hasil += angka
        angka = angka * 10 + angka_terakhir
    return hasil
```

```python
deret_bilangan(6, 7)
```

```python
#Soal 5
def deret_bintang(n):
    hasil = []
    #Naik
    for i in range(1, n+1):
        hasil.append('*' * i)
    #Turun
    for i in range(n-1, 0, -1):
        hasil.append('*' * i)
    return '\n'.join(hasil)
```

```python
print(deret_bintang(6))
```

```python
#Soal 6
def print_alternate_pattern(rows):
    hasil = 1
    for i in range(1, rows+1):
        if i % 2 != 0:
            for j in range(hasil, hasil + i):
                print(j, end=' ')
            print()
        else:
            for k in range(hasil + i-1, hasil-1,-1): #i = 2, #(3,1,-1)
                print(k, end = ' ')
            print()
        hasil += i
```

```python
print_alternate_pattern(4)
```
