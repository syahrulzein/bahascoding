```python
#Soal 1
def pengulangan_genap(end):
    start = 2
    hasil = []
    while(start <= end):
        hasil.append(start)
        start = start + 2
    return hasil
```

```python
pengulangan_genap(9)
```

```python
#Soal 2
def jumlah_ganjil_genap(start,end):
    ganjil=0
    genap=0
    for i in range(start,end+1):
        if i % 2 == 0:
            genap=genap + i
        else:
            ganjil=ganjil + i
    hasil = 'Jumlah Bilangan Ganjil =',ganjil,',Genap = ',genap
    return hasil
```

```python
jumlah_ganjil_genap(1,3)
```

```python
#Soal 3
def cetak_ori(num):
    hasil = []
    for _ in range(num):
        hasil.append("Hello World!")
    return "\n".join(hasil)
```

```python
print(cetak_ori(3))
```

```python
def cetak_angga(num):
    i = 0
    while i < num:
        i += 1
        hasil = print('Hello World')
    return hasil
```

```python
cetak_angga(3)
```

```python
def cetak_bagas(num):
    i = 0
    while i < num:
        print('Hello World!!!')
        i += 1
```

```python
#Soal 4
def bintang(num):
    hasil = []
    for i in range(1, num + 1):
        baris = ''
        for j in range(1, i + 1):
            baris += '*'
        hasil.append(baris)
    return "\n".join(hasil)
```

```python
print(bintang(5))
```

```python
#Soal 5
def angka_terbalik(start,end):
    for i in range(start, end-1, -1):
        for j in range(1, i + 1):
            print(i, end=" ")
        print()
```

```python
angka_terbalik(5,1)
```

```python
def angka_kebalik_bagas(n):
    baris = n
    while baris >= 1:
        print((str(baris)+' ')*baris)
        baris -= 1
```

```python
angka_kebalik_bagas(5)
```

```python
def perkalian_icad(num):
    depan = num // 10
    belakang = num % 10
    hasil = depan * belakang
    return hasil
```

```python
perkalian_icad(234)
```

```python
#Soal 6
def perkalian_dua_angka(num):
    p = 1
    while (num):
        r = num % 10
        p = p * r
        num = num // 10
    return p
```

```python
perkalian_dua_angka(2345)
```

```python
def perkalian_dwi(num):
    hasil = 1
    for i in str(num):
        hasil = hasil * int(i)
    return hasil 
```

```python
perkalian_dwi(2345)
```

```python
def faktorial_icad(num):
    hasil = 1
    for i in range(num):
        hasil = hasil * num
        num = num - 1
    return hasil
```

```python
faktorial_icad(3)
```

```python
def faktorial_bagas(num):
    hasil = 1
    while num > 0:
        hasil *= num
        num -= 1
    return hasil
```

```python
faktorial_bagas(5)
```

```python
#Soal 7
def faktorial(num):
    hasil = 1
    for i in range(1, num + 1):
        hasil *= i
    return hasil
```

```python
faktorial(5)
```
